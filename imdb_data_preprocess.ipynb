```python

from langchain_chroma import Chroma
from langchain_openai import AzureOpenAIEmbeddings
embeddings=AzureOpenAIEmbeddings(model="text-embedding-ada-002" , chunk_size=1, openai_api_type="azure", azure_endpoint=os.environ['AZURE_OPENAI_API_EMBEDDINGS_ENDPOINT'])
# Split documents into chunks for embedding
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_docs = text_splitter.split_documents(docs)
vector_store = Chroma(
    collection_name="example_collection",
    embedding_function=embeddings,
    persist_directory="./imdb_db",  # Where to save data locally, remove if not necessary
)
```

```python
BATCH_SIZE = 100 # Define your desired batch size

print(f"Starting to add documents to Chroma in batches of {BATCH_SIZE}...")

for i in range(0, len(split_docs), BATCH_SIZE):
    batch = split_docs[i:i + BATCH_SIZE]
    print(f"Adding batch {int(i/BATCH_SIZE) + 1} of {len(batch)} documents...")
    vector_store.add_documents(batch)
    print(f"Batch {int(i/BATCH_SIZE) + 1} added. Total documents processed: {i + len(batch)}")
    
print("\nAll documents added to Chroma DB.")
print(f"Current number of documents in collection: {vector_store._collection.count()}")
    
```
